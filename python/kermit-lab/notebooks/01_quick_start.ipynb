## Plot inline

Every plot returns a `matplotlib.figure.Figure`. `dist` works with any time-metric data, so it's the safest first plot. Other shapes have axis requirements:

- `kl.scaling(df)` — needs ≥2 distinct `tuples` values across reports.
- `kl.bar_time(df, query=…)` — pick one of the queries present in `df`.
- `kl.bar_space(df)` — needs reports with space-metric measurements.
- `kl.tradeoff(df)` — needs reports with both time and space metrics.
- `kl.bar_queries(df, ds=…, algo=…)` — needs ≥2 queries for the chosen pair.

See the other notebooks for worked examples of each.
