# WatDiv: per-query bars

Compare each WatDiv stress query's iteration time, one figure per `(data_structure, algorithm)` pair.

**Requires** `bench-runs/watdiv-stress-*.json` files with ≥2 distinct queries per `(data_structure, algorithm)` pair. The WatDiv suite is split into 12 YAMLs (`watdiv-stress-{100,1000}-{warmup,test-1..5}`); run them with e.g.:

```sh
for s in warmup test-1 test-2 test-3 test-4 test-5; do
  cargo run -p kermit -- bench fetch watdiv-stress-100-$s
  cargo run -p kermit -- bench --report-json bench-runs/watdiv-stress-100-$s.json \
    run watdiv-stress-100-$s -i tree-trie -a leapfrog-triejoin
done
```

Note: `--report-json` is a flag on the parent `bench` command, so it must come *before* `run`.

````python
import kermit_lab as kl

PATHS = '../../../bench-runs/watdiv-stress-*.json'
CRITERION_ROOT = '../../../target/criterion'

kl.apply_style()
df = kl.load(PATHS, criterion_root=CRITERION_ROOT)
df[['benchmark', 'query', 'data_structure', 'algorithm']].drop_duplicates().head()
````

````python
pairs = df[['data_structure', 'algorithm']].dropna().drop_duplicates().values.tolist()
for ds, algo in pairs:
    fig = kl.bar_queries(df, ds=ds, algo=algo)
    fig.suptitle(f'{ds} / {algo}')
    display(fig)
````
