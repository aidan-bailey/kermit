# WatDiv: per-query bars

Compare each WatDiv stress query's iteration time, one figure per `(data_structure, algorithm)` pair.

**Requires** `bench-runs/watdiv-stress-*.json` files with ≥2 distinct queries per `(data_structure, algorithm)` pair. Generate with `kermit bench run watdiv-stress-100 -i tree-trie -a leapfrog-triejoin --report-json bench-runs/watdiv-stress-100.json` (and similarly for other scales / DS / algos).

```python
import kermit_lab as kl

kl.apply_style()
df = kl.load('../../bench-runs/watdiv-stress-*.json', criterion_root='../../target/criterion')
df[['benchmark', 'query', 'data_structure', 'algorithm']].drop_duplicates().head()
```

```python
pairs = df[['data_structure', 'algorithm']].dropna().drop_duplicates().values.tolist()
for ds, algo in pairs:
    fig = kl.bar_queries(df, ds=ds, algo=algo)
    fig.suptitle(f'{ds} / {algo}')
    display(fig)
```
