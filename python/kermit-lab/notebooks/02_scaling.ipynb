# Scaling analysis

Plot mean iteration time vs `tuples` for each `(data_structure, algorithm)`, then drop into a pivot table for the raw numbers.

**Requires** bench runs at ≥2 distinct `tuples` values (e.g. the Oxford uniform suite at multiple scales). Single-scale data will raise `InsufficientAxesError`.

```python
import kermit_lab as kl

kl.apply_style()
df = kl.load('../../bench-runs/*.json', criterion_root='../../target/criterion')
print(f"distinct 'tuples' values: {sorted(df.tuples.dropna().unique().tolist())}")
```

```python
kl.scaling(df, phase='iteration')
```

## Pivot: mean iteration time by (DS, scale)

```python
iter_time = df[(df.metric == 'time') & (df.phase == 'iteration')]
kl.summary(iter_time, rows='data_structure', cols='tuples', value='mean_ns')
```
