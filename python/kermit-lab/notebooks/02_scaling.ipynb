# Scaling analysis

Plot mean iteration time vs `tuples` for each `(data_structure, algorithm)`, then drop into a pivot table for the raw numbers.

```python
import glob
import kermit_lab as kl

kl.apply_style()
df = kl.load(sorted(glob.glob('../../bench-runs/*.json')), criterion_root='../../target/criterion')
```

```python
kl.scaling(df, phase='iteration')
```

## Pivot: mean iteration time by (DS, scale)

```python
iter_time = df[(df.metric == 'time') & (df.phase == 'iteration')]
kl.summary(iter_time, rows='data_structure', cols='tuples', value='mean_ns')
```
