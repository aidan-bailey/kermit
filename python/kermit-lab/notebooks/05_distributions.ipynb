# Per-iter distributions and significance

Violin plot of raw Criterion samples per `(DS, algorithm)`, then a Mann-Whitney U test asking "are these distributions distinguishable?"

Violin works with any time-metric data. The Mann-Whitney section additionally **requires** ≥2 distinct `data_structure` values; the notebook auto-picks the first two it finds.

```python
import kermit_lab as kl

kl.apply_style()
df = kl.load('../../bench-runs/*.json', criterion_root='../../target/criterion')
samples = kl.load_samples('../../bench-runs/*.json', criterion_root='../../target/criterion')
```

```python
kl.dist(df, df_samples=samples)
```

## Mann-Whitney U: pairwise distribution test

```python
ds_values = sorted(df.data_structure.dropna().unique().tolist())
assert len(ds_values) >= 2, f'need ≥2 data_structures for Mann-Whitney; got {ds_values}'
A, B = ds_values[0], ds_values[1]

labelled = samples.merge(
    df[['criterion_group', 'criterion_function', 'data_structure', 'phase']],
    on=['criterion_group', 'criterion_function'],
)
a = labelled.query(f"data_structure == @A and phase == 'iteration'").per_iter_ns.to_numpy()
b = labelled.query(f"data_structure == @B and phase == 'iteration'").per_iter_ns.to_numpy()
u, p = kl.mannwhitney_u(a, b)
print(f'{A} vs {B}:  U = {u:.0f}, p = {p:.4g}')
```
